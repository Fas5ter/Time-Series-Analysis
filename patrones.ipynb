# Problema de identificación de patrones en datos de ventas mensuales

## Descripción del problema
En una empresa de venta al por menor, se ha recopilado un conjunto de datos que registra las ventas mensuales de varios productos durante un periodo de varios años. Tu tarea es analizar estos datos para identificar posibles patrones o tendencias en las ventas mensuales.


## Parte 1: Generación de Datos
Utiliza un programa en Python para generar datos simulados que representen las ventas mensuales de varios productos a lo largo de un período de tiempo. Los datos deben incluir al menos 3 productos diferentes y abarcar un periodo de al menos 3 años.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# 11, 13, 17
# Crear rango de fechas mensuales para tres años
meses = pd.date_range(start='2022-01-01', periods=36, freq='ME')
np.random.seed(11)
producto1 = np.random.normal(loc=600, scale=200, size=36)
np.random.seed(13)
producto2 = np.random.normal(loc=400, scale=150, size=36)
np.random.seed(17)
producto3 = np.random.normal(loc=200, scale=250, size=36)

df_ventasp1 = pd.DataFrame({'Ventas': producto1}, index=meses) # Usar meses como indice
df_ventasp2 = pd.DataFrame({'Ventas': producto2}, index=meses) 
df_ventasp3 = pd.DataFrame({'Ventas': producto3}, index=meses) 
df_ventas = pd.DataFrame({'VentasP1': producto1, 'VentasP2': producto2, 'VentasP3': producto3}, index=meses) # Usar meses como indice
df_ventas
```

## Parte 2: Análisis de Datos
Una vez que hayas generado los datos, realiza un análisis para identificar posibles patrones en las ventas mensuales. Algunas preguntas que podrías explorar incluyen:
- ¿Hay algún patrón estacional en las ventas de ciertos productos?
- ¿Se observa alguna tendencia de crecimiento o decrecimiento en las ventas a lo largo del tiempo?
- ¿Existen meses específicos en los que las ventas tienden a ser más altas o más bajas?

```python
import matplotlib.dates as mdates
# Descomposición de la serie temporal
result = seasonal_decompose(df_ventas['VentasP1'], model='additive')
result2 = seasonal_decompose(df_ventas['VentasP2'], model='additive')
result3 = seasonal_decompose(df_ventas['VentasP3'], model='additive')
fig = result.plot()
fig2 = result2.plot()
fig3 = result3.plot()
fig.set_size_inches(10, 8) # Ajustar tamaño de la figura para mejor visualización (ancho, alto)
fig2.set_size_inches(10, 8)
fig3.set_size_inches(10, 8)
plt.show()
```

```python
# import matplotlib.pyplot as plt
import matplotlib.dates as mdates
# from statsmodels.tsa.seasonal import seasonal_decompose

# Descomposición de la serie temporal
result = seasonal_decompose(df_ventas['VentasP1'], model='additive')
result2 = seasonal_decompose(df_ventas['VentasP2'], model='additive')
result3 = seasonal_decompose(df_ventas['VentasP3'], model='additive')
```
