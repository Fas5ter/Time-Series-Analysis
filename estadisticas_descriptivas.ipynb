## Contexto: 
Una empresa desea realizar un análisis estadístico de los salarios anuales de sus
empleados. El propósito de este análisis es obtener una mejor comprensión de la
distribución de los ingresos entre los empleados, lo que permitirá a la empresa tomar
decisiones informadas respecto a la equidad salarial y la estructura de compensaciones.

## Objetivo: 
Como parte de un proyecto de análisis de datos, se te ha asignado la tarea de
calcular las estadísticas descriptivas básicas de los salarios anuales en la empresa.
Específicamente, deberás calcular la media, mediana, moda, varianza y desviación estándar
de los salarios. Además, deberás interpretar estas estadísticas para discutir la equidad
salarial y la dispersión de los salarios.

## 1. Generar Datos: 
Utiliza el siguiente código en Python para generar una muestra
aleatoria de salarios anuales. Esta muestra simulará los salarios anuales de los
empleados de la empresa.

```python
import numpy as np
# Configurar la semilla del generador aleatorio para reproducibilidad
np.random.seed(13)
# Generar datos de salarios anuales (simulados)
salarios = np.random.normal(loc=50000, scale=15000, size=250) 
salarios = np.round(salarios, -3)
# media=50000, desv_std=15000, n=250
# Asegurarse de que todos los salarios sean positivos
salarios = np.abs(salarios)
salarios
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(10,6))
# plt.hist(salarios, bins=10, color='skyblue', edgecolor='black')
# plt.bar(x=np.arange(len(salarios)), height=salarios)
plt.plot(salarios, 'o', label='Salarios') # Puntos de los salarios
plt.title('Distribución de salarios anuales')
plt.ylabel('Salario anual')
plt.xlabel("Empleados")
plt.show()
```

## 2. Calcular Estadísticas Descriptivas: 
– Calcula la media, mediana, moda, varianza y desviación estándar de los
salarios generados. – Puedes usar las librerías numpy para media, mediana, varianza y desviación
estándar, y scipy.stats o statistics para la moda.

```python
import scipy.stats as stats
import statistics
media  = np.mean(salarios)
mediana = np.median(salarios)
moda = stats.mode(salarios)
# moda = statistics.mode(salarios)
varianza = np.var(salarios)
destd =np.std(salarios, ddof=1)
print(f"Media: {media}")
print(f"Mediana: {mediana}")
print(f"Moda: {moda[0]} con {moda[1]} ocurrencias")
print(f"Varianza: {varianza}")
print(f"Desviacion estandar: {destd}")
# stats.describe(salarios)
```

```python
from collections import Counter
# Utilizar Counter para contar las frecuencias de cada elemento en el conjunto de datos
frecuencias = Counter(salarios)
# Obtener la frecuencia máxima
maxima_frecuencia = max(frecuencias.values())
# Filtrar los elementos que tienen esta frecuencia máxima
modas = [num for num, freq in frecuencias.items() if freq ==
maxima_frecuencia]
# Imprimir las modas y su frecuencia común
print("Modas:", modas)
print("Frecuencia de las modas:", maxima_frecuencia)
```
